```python
import gradio as gr
import re
```

```python
def extract_text(image_file):
    if not os.path.exists(image_file):
        return "Image file not found."

    try:
        res = model.chat(tokenizer, image_file, ocr_type='ocr')
        print("Model output:", res)

        extracted_text = res if isinstance(res, str) else res.get('text', "")
        return extracted_text
    except Exception as e:
        return f"Error during OCR processing: {str(e)}"

def highlight_keywords(extracted_text, keyword):
    if not extracted_text:
        return "No text extracted to highlight."

    highlighted_text = re.sub(
        f"({re.escape(keyword)})",
        r"<mark>\1</mark>",
        extracted_text,
        flags=re.IGNORECASE
    )
    return highlighted_text

def process_image(image_file, keyword):
    try:
        print("Received image file:", image_file)
        extracted_text = extract_text(image_file)
        print("Extracted text:", extracted_text)
        highlighted_text = highlight_keywords(extracted_text, keyword)
        return highlighted_text
    except Exception as e:
        return f"Error: {str(e)}"

interface = gr.Interface(
    fn=process_image,
    inputs=[
        gr.Image(type="filepath", label="Upload Image (JPEG/PNG)"),
        gr.Textbox(label="Enter Keyword to Search", placeholder="Keyword")
    ],
    outputs=gr.HTML(label="Extracted Text with Highlights"),
    title="OCR Web Application",
    description="Upload an image to perform OCR and search for keywords."
)

interface.launch()
```
